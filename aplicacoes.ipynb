# Aula 01 - Manipulação de arquivos

- abrir arquivo -> função open()
- necessário sempre fechar a conexão com o arquivo após o processo

- readline -> ler apenas uma linha por vez
--> com arquivos no python, é criado um cursor para leitura do arquivo e os comandos vão sempre ser executados a partir do local onde o cursor do python se localiza

```python
arquivo = open('dom_casmurro_cap_1.txt', 'r', encoding='utf-8')
texto = arquivo.read()
print(texto)
arquivo.close()
```

```python
arquivo = open('dom_casmurro_cap_1.txt', 'r', encoding='utf-8')
linha = arquivo.readline()
while linha != '':
    print(linha, end='')
    linha = arquivo.readline()
arquivo.close()
```

```python
arquivo = open('dom_casmurro_cap_1.txt', 'r', encoding='utf-8')
for linha in arquivo:
    print(linha, end='')
arquivo.close()
```

- with -> abre e fecha arquivos no python

```python
with open('dom_casmurro_cap_1.txt', 'r', encoding='utf-8') as arquivo:
    texto = arquivo.read()
    print(texto)
```

- criar um novo arquivo:
    - with
    - argumento 'w' cria ou sobrescreve um arquivo
    - arquivo.write()

```python
with open('arquivo_teste.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write('Essa é uma linha que eu escrevi usando Python\n')
    arquivo.write('Essa é a segunda linha que eu escrevi usando Python\n')
```

```python
with open('arquivo_teste.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read(), end='')
```

- Adicionar um texto num arquivo:
    - argumento 'a'

```python
with open('arquivo_teste.txt', 'a', encoding='utf-8') as arquivo:
    arquivo.write('Essa é a terceira linha que eu escrevi usando Python\n')
```

# Aula 02 - Arquivos CSV

- csv - comma separated values
    - muito utilizado para criação e exportação de planilhas

- biblioteca csv nativa do python (só é necessário importar)

- next() - pula uma das interações, por exemplo o header

```python
import csv
```

```python
with open('brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    for linha in leitor:
        print(linha)
```

```python
with open('brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    header = next(leitor)
    for linha in leitor:
        if float(linha[2]) > 1:
            print(linha)
```

- Sem usar a biblioteca CSV (com apenas comandos básicos do python)

```python
with open('brasil_covid.csv', 'r') as csv_file:
    linhas = csv_file.read()
    linhas = linhas.split('\n')
    for linha in linhas:
        linha = linha.split(',')
        print(linha)
```

Escrita no csv:
 

```python
with open('users.csv', 'w', encoding='utf-8', newline='') as arquivo_users:
    escritor = csv.writer(arquivo_users)
    escritor.writerow(['nome', 'sobrenome', 'email', 'genero'])
    escritor.writerow(['Pietro', 'Ribeiro', 'pietro@email.com', 'Masculino'])
```

```python
from email.header import Header


header = ['nome', 'sobrenome']
dados = []
opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n')
while opt != '0':
    nome = input('Qual seu nome?')
    sobrenome = input('Qual seu sobrenome?')
    dados.append([nome,sobrenome])
    opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n')

print(dados)

with open('users.csv', 'w', newline='') as arquivo_csv:
    writer = csv.writer(arquivo_csv)
    writer.writerow(header)
    writer.writerows(dados)

with open('users.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        print(row)
```

```python
url = 'https://api.exchangerate-api.com/v6/latest'

req = requests.get(url)

print(req.status_code)
```

```python
dados = req.json()

print(dados)
```
